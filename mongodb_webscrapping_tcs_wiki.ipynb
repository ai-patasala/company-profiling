```python
# import necessary libraries
from bs4 import BeautifulSoup
import requests
import re
  
  
# function to extract html document from given url
def getHTMLdocument(url):
      
    # request for HTML document of given url
    response = requests.get(url)
      
    # response will be provided in JSON format
    return response.text
  
    
# assign required credentials
# assign URL
url_to_scrape = "https://en.wikipedia.org/wiki/Tata_Consultancy_Services"

# create document
html_document = getHTMLdocument(url_to_scrape)
#len(html_document)

# create soap object
soup = BeautifulSoup(html_document, 'html.parser')
soup.prettify()
```

```python
#for getting all labels with its class
#soup.findAll(class_="infobox-label")

#get table data
table = soup.findAll(class_="infobox vcard")
tr = table[0].findAll('tr')
len(tr)
```

```python
def founder_function():
    #global founder
    for i in tr:
            for m in i.findAll('th'):        
                for j in i.findAll('td'):
                    try:
                        for k in j.findAll('a'):
                            if m.string == 'Founder':                        
                                #print(k.string)
                                founder = k.string
                                return founder
                    except:
                        pass
```

```python
def founded_function():
    for i in tr:
        for m in i.findAll('th'):        
            for j in i.findAll('td'):
                try:
                    if m.string == 'ISIN':
                        #print(j.string)
                        founded = j.string
                        return founded
                except:
                    pass 
```

```python
def revenue_function():
    for i in tr:
        for m in i.findAll('th'):        
            for j in i.findAll('td'):
                try:
                    #for k in j.findAll('span'):                    
                    if m.string == 'Revenue':
                        #print(j.contents[2])
                        #print(j.contents[3])
                        #print(j.contents[5])
                        data = j.contents[3]+j.contents[5]
                        return data
                except:
                    pass 
```

```python
def noofemp_function():
    for i in tr:
        for m in i.findAll('th'):        
            for j in i.findAll('td'):
                try:
                    #for k in j.findAll('span'):                    
                    if m.string == 'Number of employees':                        
                        #print(j.contents[0]+j.contents[2])
                        data = j.contents[0]+j.contents[2]
                        return data
                except:
                    pass  
```

```python
def website_function():
    for i in tr:
        for m in i.findAll('th'):        
            for j in i.findAll('td'):
                try:
                    for k in j.findAll('a'):                    
                        if m.string == 'Website': 
                            #print(k.contents[0]+k.contents[2]+k.contents[4])
                            website = k.contents[0]+k.contents[2]+k.contents[4]
                           #print(website)
                            return website
                except:
                    pass 
```

```python
def industries_function():
    for i in tr:
        for m in i.findAll('th'):        
            for j in i.findAll('td'):
                try:
                    for k in j.findAll('a'):                    
                        if m.string == 'Industry': 
                            #print(k.contents[0])
                            industry = k.contents[0]
                            return industry
                except:
                    pass  
```

```python
def getcompanyname_function():
    company = soup.find('h1', {'class' : 'firstHeading'})
    #print(company.contents[0])
    companyname = company.contents[0]
    return companyname
```

```python
founder = founder_function()
print(founder)

founded = founded_function()
print(founded)

revenue = revenue_function()
print(revenue)

noemp = noofemp_function()
print(noemp)

website = website_function()
print(website)

industry = industries_function()
print(industry)

companyname = getcompanyname_function()
print(companyname)
```
