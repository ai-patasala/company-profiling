```python
# import necessary libraries
from bs4 import BeautifulSoup
import requests
import re
  
  
# function to extract html document from given url
def getHTMLdocument(url):
      
    # request for HTML document of given url
    response = requests.get(url)
      
    # response will be provided in JSON format
    return response.text
  
    
# assign required credentials
# assign URL
url_to_scrape = "https://markets.ft.com/data/equities/tearsheet/profile?s=TCS:NSI"

# create document
html_document = getHTMLdocument(url_to_scrape)
#len(html_document)

# create soap object
soup = BeautifulSoup(html_document, 'html.parser')
soup.prettify()
```

```python
#for getting all labels with its class
#soup.findAll(class_="infobox-label")

#get table data
#data = soup.findAll(class_="infoTableItem__7381420557")
mydivs = soup.findAll("div", class_="mod-tearsheet-profile")
#print(mydivs)
ul = mydivs[0].findAll('ul')
len(ul)
```

```python
for i in ul:
    for m in i.findAll('span'):        
        try:
            print(m.string)
        except:
            pass           
```

```python
def revenueft_function():
    for i in ul: 
        for k in i.findAll('span'): 
            try:
                revenueheader = k.string
                if k.string == 'Revenue in INR (TTM)':
                    revenueitem = revenueheader.next()                
                    return revenueitem[0].string
            except:
                pass 
```

```python
def netincomeft_function():
    for i in ul: 
        for k in i.findAll('span'): 
            try:
                netincomeheader = k.string
                if k.string == 'Net income in INR':
                    netincomeitem = netincomeheader.next()                
                    return netincomeitem[0].string
            except:
                pass 
```

```python
#def incorporatedft_function():
for i in ul: 
    for k in i.findAll('span'): 
        try:
            incorporatedheader = k.string
            if k.string == 'Incorporated':
                incorporateditem = incorporatedheader.next()                
                print(incorporateditem)
        except:
            pass 
```

```python
def empft_function():
    for i in ul: 
        for k in i.findAll('span'): 
            try:
                empheader = k.string
                if k.string == 'Employees':
                    empitem = empheader.next()                
                    return empitem.content
            except:
                pass 
```

```python
revenue = revenueft_function()
print(revenue)

netincome = netincomeft_function()
print(netincome)

incorporated = incorporatedft_function()
print(incorporated)

emp = empft_function()
print(emp)
```
