```python
def keypersons_function():
    keyp = []
    res = soup.find_all('script')
    #json_object = json.loads(res.contents[0])
    secscript = res[1]
    json_object = json.loads(secscript.contents[0])
    employees = json_object['employee']

    for emp in employees:
        #for l in t:
        #print(emp['name'])
        m = emp['name']   
        keyp.append(m)
    return keyp
```

```python
def emp_function():
    res = soup.find_all('script')
    #json_object = json.loads(res.contents[0])
    secscript = res[1]
    json_object = json.loads(secscript.contents[0])
    return json_object['numberOfEmployees']
```

```python
def website_function():   
    secscript = res[1]
    json_object = json.loads(secscript.contents[0])
    return json_object['url']
```

```python
def telephone_function():   
    secscript = res[1]
    json_object = json.loads(secscript.contents[0])
    return json_object['telephone']
```

```python
def foundingDate_function():   
    secscript = res[1]
    json_object = json.loads(secscript.contents[0])
    return json_object['foundingDate']
```

```python
def address_function():   
    secscript = res[1]
    json_object = json.loads(secscript.contents[0])
    return json_object['address']
```

```python
keypersons = keypersons_function()
print(keypersons)


numberOfEmployees = emp_function()
print(numberOfEmployees)

website = website_function()
print(website)

telephone = telephone_function()
print(telephone)

foundingDate = foundingDate_function()
print(foundingDate)

address = address_function()
print(address)
```
